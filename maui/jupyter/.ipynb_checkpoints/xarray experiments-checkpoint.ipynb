```python
%matplotlib inline
import xarray as xr
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
net_infil = xr.open_dataset('../output/maui_net_infiltration_2001_2002__210_by_316.nc')
net_infil_df = net_infil.to_dataframe
net_infil
```

```python
mask_water = 2 * np.zeros( ( net_infil.dims['y'], net_infil.dims['x'] ) ) * np.isnan( net_infil.net_infiltration.isel(time=0))
mask_land = 1 * np.ones( ( net_infil.dims['y'], net_infil.dims['x'] ) ) * np.isfinite( net_infil.net_infiltration.isel(time=0))
mask_array = mask_water + mask_land
mask_array.plot()
```

```python
net_infil.coords['mask'] = (('y','x'), mask_array )
net_infil
```

```python
net_infil.net_infiltration[179].plot()
```

```python
net_infil
```

```python
june_net_infil = net_infil.net_infiltration.sel(time=slice('2001-07-01','2001-07-31'))
june_sum = june_net_infil.where( mask_array == 1 ).sum(dim='time')
```

```python
june_sum.where(june_sum.mask == 1).plot()
```

```python
net_infil_bymonth = net_infil['net_infiltration'].where( net_infil.mask==1).groupby('time.month').sum(dim='time')
net_infil_bymonth.where( net_infil.mask==1).groupby('month').mean()
```

```python
net_infil_byseason = net_infil['net_infiltration'].where( net_infil.mask==1).groupby('time.season').sum(dim='time')
```
